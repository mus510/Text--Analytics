```python
# ====================================================
# CISB5123 Text Analytics - Lab Assignment 1
# Name: Hashem Musaab
# Student ID: SW01084071
#
# This script scrapes 5 pages of eBay product reviews
# and extracts:
# - Reviewer Name
# - Review Date
# - Review Content
# Then saves the data to CSV.
# ====================================================

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import csv
import time

def setup_driver():
    """
    Set up Microsoft Edge WebDriver
    """

    edge_options = Options()
    edge_options.add_argument("--start-maximized")

    # Change this path if yours is different
    service = Service(r"D:\user\Downloads\edgedriver_win64\msedgedriver.exe")
    driver = webdriver.Edge(service=service, options=edge_options)

    return driver

def scrape_reviews(driver, urls):
    """
    Loop through each review page URL,
    extract review data, and return results
    """

    reviews_data = []

    for url in urls:
        driver.get(url)
        time.sleep(4)  # Wait for the page to load fully

        review_blocks = driver.find_elements(By.CLASS_NAME, "ebay-review-section")

        for review in review_blocks:

            try:
                reviewer = review.find_element(By.CLASS_NAME, "review-item-author").text
            except:
                reviewer = "Not Available"

            try:
                date = review.find_element(By.CLASS_NAME, "review-item-date").text
            except:
                date = "Not Available"

            try:
                content = review.find_element(By.CLASS_NAME, "review-item-content").text
            except:
                content = "Not Available"

            reviews_data.append({
                "Reviewer Name": reviewer,
                "Review Date": date,
                "Review Content": content
            })

    return reviews_data

def save_to_csv(data, filename):
    if len(data) == 0:
        print("No reviews extracted.")
        return

    keys = data[0].keys()

    with open(filename, "w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()
        writer.writerows(data)

# ====================== MAIN ======================

review_urls = [
    "https://www.ebay.com/urw/Inkbird-ITC-308-Thermostat-Temperature-Controller/product-reviews/2296084805?pgn=1",
    "https://www.ebay.com/urw/Inkbird-ITC-308-Thermostat-Temperature-Controller/product-reviews/2296084805?pgn=2",
    "https://www.ebay.com/urw/Inkbird-ITC-308-Thermostat-Temperature-Controller/product-reviews/2296084805?pgn=3",
    "https://www.ebay.com/urw/Inkbird-ITC-308-Thermostat-Temperature-Controller/product-reviews/2296084805?pgn=4",
    "https://www.ebay.com/urw/Inkbird-ITC-308-Thermostat-Temperature-Controller/product-reviews/2296084805?pgn=5"
]

driver = setup_driver()

all_reviews = scrape_reviews(driver, review_urls)

save_to_csv(all_reviews, "eBay_Reviews.csv")

driver.quit()

print("Scraping completed. File saved as eBay_Reviews.csv")
```
